# Loading the MNIST Dataset

Importing Packages

```python
import torch
import torch.nn as nn
from torch.optim import Adam
from torchvision import datasets
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor, Compose, Normalize, Lambda
from matplotlib import pyplot as plt
```

Device selection

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

Hyperparameters

```python
batch_size = 5000
```

Loading the dataset

```python
train_data = datasets.MNIST(root='./data',
                            train=True,
                            download=True, 
                            transform=Compose([ToTensor(),Normalize((0.1307,), (0.3081,)), Lambda(lambda x : torch.flatten(x))]))
test_data = datasets.MNIST(root='./data',
                           train=False,
                            download=True, 
                            transform=Compose([ToTensor(),Normalize((0.1307,), (0.3081,)), Lambda(lambda x : torch.flatten(x))]))


x_train, y_train = train_data.data.to(device), train_data.targets.to(device)
x_test, y_test = test_data.data.to(device), test_data.targets.to(device)
```
